# Import the KDTree class from the file

```python
from KDTree import KDTree
import matplotlib.pyplot as plt

import random
%matplotlib inline
```

## Create a tree and give it a bunch of points

```python
tree = KDTree()
goal_node = None
for i in range(50):
    x = random.uniform(-20, 20.0)
    y = random.uniform(-20, 20.0)
    tree.insert((x,y))
    if i == 49:
        x = random.uniform(-20, 20.0)
        y = random.uniform(-20, 20.0)
        goal_node =  tree.insert((x,y))
        goal_node.set_goal_path()
        
        
```

### This is NOT an rrt plot, but if you insert poinmts using the actual rrt algorithm and not randomly...

```python
fig, ax = plt.subplots()
tree.PlotTree(fig, ax)
```
